```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained model and tokenizer
model_name = "gpt2"  # You can change this to other GPT-2 models like "gpt2-medium", "gpt2-large", etc.
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Encode input context to get token ids
input_text = "The purpose of life is"  # Example input text
input_ids = tokenizer.encode(input_text, return_tensors="pt")

# Greedy decoding
greedy_output = model.generate(input_ids, max_length=50, num_return_sequences=1, do_sample=False)
print("Greedy decoding output:", tokenizer.decode(greedy_output[0], skip_special_tokens=True))
```

```python
# Beam search decoding
beam_output = model.generate(
    input_ids,
    max_length=50,
    num_beams=5,
    num_return_sequences=1,
    early_stopping=True,
    no_repeat_ngram_size=2
)
print("Beam search decoding output:", tokenizer.decode(beam_output[0], skip_special_tokens=True))
```
